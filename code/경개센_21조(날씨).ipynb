```python
import pandas as pd
pd.set_option('display.width', 200)
```

```python
import matplotlib.pyplot as plt
```

```python
def set_col_idx(df) :
    df.drop(['Month', 'Year'], axis=1, inplace=True)
    df.index = ['avg_high', 'avg_low', 'avg_rainfall', 'avg_raindays', 'avg_hum', 'mean_sun_hours']
    for col in df.columns :
        df[col] = df[col].apply(lambda x : x.split()[0])
```

```python
url_mel = "https://en.wikipedia.org/wiki/Climate_of_Melbourne"
df_mel = pd.read_html(url_mel)[0]
df_mel.columns = df_mel.columns.droplevel(0)
df_mel.drop([0, 2, 4, 9, 10, 11], axis=0, inplace=True)
set_col_idx(df_mel)
df_mel = df_mel.astype('float')
print(df_mel)
```

```python
url_syd = "https://en.wikipedia.org/wiki/Climate_of_Sydney"
df_syd = pd.read_html(url_syd)[4]
df_syd.columns = df_syd.columns.droplevel(0)
df_syd.drop([0, 3, 7, 9], axis=0, inplace=True)
set_col_idx(df_syd)
df_syd = df_syd.astype('float')
print(df_syd)
```

```python
url_ade = "https://en.wikipedia.org/wiki/Climate_of_Adelaide"
df_ade = pd.read_html(url_ade)[0]
df_ade.columns = df_ade.columns.droplevel(0)
df_ade.drop([0, 1, 4, 5, 10], axis=0, inplace=True)
set_col_idx(df_ade)
df_ade = df_ade.astype('float')
print(df_ade)
```

```python
url_bri = "https://en.wikipedia.org/wiki/Climate_of_Brisbane"
# 0 8 10 12
df_bri = pd.DataFrame()
for i in [0, 8, 10, 12] :
    df = pd.read_html(url_bri)[i][:-1]
    df.columns = df.columns.droplevel(0)
    df_bri = pd.concat([df_bri, df], axis=0)
set_col_idx(df_bri)
df_bri = df_bri.astype('float')
print(df_bri)
```

```python
url_per = "https://en.wikipedia.org/wiki/Climate_of_Perth"
df_per = pd.read_html(url_per)[5]
df_per.columns = df_per.columns.droplevel(0)
df_per.drop([0, 1, 4, 5, 9, 11], axis=0, inplace=True)
set_col_idx(df_per)
for mon in [0, 2, 4, 6, 7, 9, 11] :
    df_per.iloc[5, mon] = float(df_per.iloc[5, mon]) * 31
for mon in [3, 5, 8, 10] :
    df_per.iloc[5, mon] = float(df_per.iloc[5, mon]) * 30
df_per.iloc[5, 1] = float(df_per.iloc[5, 1]) * 28
df_per = df_per.astype('float')
print(df_per)
```

```python
url_seo = "https://en.wikipedia.org/wiki/Climate_of_Seoul"
df_seo = pd.read_html(url_seo)[3]
df_seo.columns = df_seo.columns.droplevel(0)
df_seo.drop([0, 2, 4, 7, 10, 11, 12, 13], axis=0, inplace=True)
set_col_idx(df_seo)
df_seo.iloc[1, 0] = -5.5
df_seo.iloc[1, 1] = -3.2
df_seo.iloc[1, -1] = -3.4
df_seo = df_seo.astype("float")
print(df_seo)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(20, 8))
plt.setp(axes, xticks=range(12), xticklabels=df_mel.columns)

df_mel.iloc[0].plot(ax=axes[0, 0])
df_mel.iloc[1].plot(ax=axes[0, 0])
axes[0, 0].set_title("Melbourne")

df_syd.iloc[0].plot(ax=axes[0, 1])
df_syd.iloc[1].plot(ax=axes[0, 1])
axes[0, 1].set_title("Sydney")

df_ade.iloc[0].plot(ax=axes[0, 2])
df_ade.iloc[1].plot(ax=axes[0, 2])
axes[0, 2].set_title("Adelaide")

df_bri.iloc[0].plot(ax=axes[1, 0])
df_bri.iloc[1].plot(ax=axes[1, 0])
axes[1, 0].set_title("Brisbane")

df_per.iloc[0].plot(ax=axes[1, 1])
df_per.iloc[1].plot(ax=axes[1, 1])
axes[1, 1].set_title("Perth")

df_seo.iloc[0].plot(ax=axes[1, 2])
df_seo.iloc[1].plot(ax=axes[1, 2])
axes[1, 2].set_title("Seoul")

handles, labels = axes[0, 0].get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center')

plt.show()
```
